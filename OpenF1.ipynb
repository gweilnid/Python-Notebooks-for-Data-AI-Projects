```python
country = "Canada"
position = 1
```

```python
import requests
import json
import pandas as pd
```

```python
url = f"https://api.openf1.org/v1/meetings?year=2025&country_name={country}"
response = requests.get(url)
obj = response.json()
df = pd.DataFrame(obj)
print(df)
meeting_key = str(obj[0]["meeting_key"])
print(f'Meeting key: {meeting_key}')
```

```python
winner_url = f"https://api.openf1.org/v1/position?meeting_key={meeting_key}"
winner_response = requests.get(winner_url)
winner_obj = winner_response.json()
df_2 = pd.DataFrame(winner_obj)
print(df_2)
```

```python
last_session = df_2["session_key"].max()
print(f"Last session: {last_session}")
df2_filtered = df_2[df_2["session_key"] == last_session]
print(df2_filtered)
```

```python
df2_filtered['date'] = pd.to_datetime(df2_filtered['date'], format='ISO8601')
df_sorted = df2_filtered.sort_values(by='date',ascending=False)
print(df_sorted)
```

```python
winner_racer = df_sorted[df_sorted['position'] == position]
print(winner_racer.iloc[0])
```

```python
names_api = "https://api.openf1.org/v1/drivers"
names_re = requests.get(names_api)
names_obj = names_re.json()
names_df = pd.DataFrame(names_obj)
print(names_df)
```

```python
final = names_df[(names_df['driver_number'] == winner_racer.iloc[0]['driver_number']) & (names_df['session_key'] == last_session)]
print(final.iloc[0]['broadcast_name'])
```
